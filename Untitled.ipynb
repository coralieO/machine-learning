```python
import pandas as pd
from langdetect import detect

df = pd.read_csv('Data_Projet/text_sentiment.csv')

df
```

```python
df.drop_duplicates()
df = df.dropna()
df
```

```python
print(df.columns)
```

```python
def detect_lang(text):
    try:
        return detect(text)
    except:
        return None

# Supprimer les colonnes que vous ne souhaitez pas conserver
colonnes_a_supprimer = ['tweet_id', 'author']  # Remplacez par les noms de vos colonnes à supprimer
df.drop(columns=colonnes_a_supprimer, inplace=True)

df
```

```python
# Appliquer la fonction à la colonne contenant le texte
df['langue_detectee'] = df['content'].apply(detect_lang)

# Filtrer le DataFrame
df_filtered = df[(df['langue_detectee'] == 'en') &  # Les lignes en anglais
                 (~df['content'].str.contains('#|@', na=False)) &  # Les lignes sans '#' ou '@'
                 (df['content'].str.len() >= 1000)].copy()  # Les lignes avec plus de 1000 caractères, et créer une copie

# Supprimer la colonne 'langue_detectee' si vous ne voulez pas la conserver
df_filtered.drop(columns=['langue_detectee'], inplace=True)

# Réindexer le DataFrame filtré
df_filtered.reset_index(drop=True, inplace=True)

# Enregistrer le DataFrame filtré dans un nouveau fichier CSV
df_filtered.to_csv("sentiment_new.csv", index=False)
```
